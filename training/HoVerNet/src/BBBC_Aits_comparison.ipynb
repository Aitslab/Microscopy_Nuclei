```python
import cv2
import matplotlib.pyplot as plt
import glob
import numpy as np

images = glob.glob('./*.png')
print(images)

for i in images:
    plt.figure(figsize = (5, 3))#open new figure dont over right next iteration
    img = cv2.imread(i)
    plt.imshow(img)
    plt.figure(figsize = (5, 3))
    print(np.histogram(img))
    plt.hist(img.ravel(), 255, [0, 255])
    plt.show
```
