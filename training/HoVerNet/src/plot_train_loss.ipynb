This notebook is for plotting HoVErNet train-validation loss during training process for 50 epochs.


```python
import pandas as pd
import matplotlib.pyplot as plt

MSE = []
VAL_MSE = []
with open ('loss/train_loss_hv_mse.txt','r') as mse:
    for k in mse:
        MSE.append(float(k))

with open ('loss/valid-hv_mse.txt','r') as v_mse:
    for k in v_mse:
        VAL_MSE.append(float(k))

plt.plot(MSE,'x-c',VAL_MSE,'.-m')
plt.xlabel('Epoch',fontsize = 16)
plt.ylabel('Mean Squared Error',fontsize = 16)
plt.legend(['Train','Validation'],fontsize=16)
plt.rc('xtick', labelsize=16) 
plt.rc('ytick', labelsize=16) 

plt.tight_layout()
plt.savefig('../../../model/HoVerNet/MSE_HOverNEt.png')
plt.savefig('../../../model/HoVerNet/MSE_HOverNEt.eps')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

DICE = []
VAL_DICE = []
with open ('loss/train_loss_np_dice.txt','r') as dice:
    for k in dice:
        DICE.append(1-float(k))

with open ('loss/valid-np_dice.txt','r') as v_dice:
    for k in v_dice:
        VAL_DICE.append(float(k))

plt.plot(DICE,'x-c',VAL_DICE,'.-m')
plt.xlabel('Epoch',fontsize = 16)
plt.ylabel('DICE',fontsize = 16)
plt.legend(['Train','Validation'],fontsize =16)
plt.rc('xtick', labelsize=16) 
plt.rc('ytick', labelsize=16) 

plt.tight_layout()
plt.savefig('../../../model/HoVerNet/DICE_HOverNEt.png')
plt.savefig('../../../model/HoVerNet/DICE_HOverNEt.eps')

```
