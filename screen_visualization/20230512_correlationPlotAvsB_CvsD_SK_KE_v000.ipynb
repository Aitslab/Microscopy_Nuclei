## Raw count 

```python
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt


A = np.load('A_count.npy')
B = np.load('B_count.npy')
C = np.load('C_count.npy')
D = np.load('D_count.npy')


sns.violinplot([A,B,C,D])


```

```python
from scipy.stats.stats import spearmanr,pearsonr,linregress


corr_c, pval_c = pearsonr(A[0],B[0])
print(corr_c,pval_c)


slope_0, intercept, r_value, p_value, std_err = linregress(A[0],B[0])
print(slope_0, intercept, r_value, p_value, std_err)

corr_s, pval_s = pearsonr(C[0],D[0])
print(corr_s,pval_s,)
slope_1, intercept_1, r_value_1, p_value_1, std_err_1 = linregress(C[0],D[0])
print(slope_1, intercept_1, r_value_1, p_value_1, std_err_1)

x = np.arange(100,2000,1)

plt.figure(figsize=(15,15))
plt.scatter(A[0],B[0], facecolors='none',edgecolors='k',s=80)
plt.text(200, 1400, 'correlation= {:.2f}'.format(r_value))
plt.text(200, 1350, 'p-value < 0.001')
plt.text(200, 1300, 'std_err= {:.2f}'.format(std_err))

plt.plot(x,slope_0*x+intercept,'r')

#plt.plot(x,x,c='g')
plt.xlim([100,2000])
plt.ylim([100,2000])
plt.xlabel('A')
plt.ylabel('B')
plt.figure(figsize=(15,15))



plt.scatter(C[0],D[0],facecolors='none',edgecolors='k',s=80)
#plt.plot(x,x,c='g')
plt.plot(x,slope_1*x+intercept_1,'r')
plt.text(200, 1400, 'correlation= {:.2f}'.format(r_value_1))
plt.text(200, 1350,'p-value < 0.001')
plt.text(200, 1300, 'std_err= {:.2f}'.format(std_err_1))
plt.xlim([100,2000])
plt.ylim([100,2000])
plt.xlabel('C')
plt.ylabel('D')

plt.rcParams.update({'font.size': 22})
```
