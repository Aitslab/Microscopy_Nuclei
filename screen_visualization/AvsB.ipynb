```python
#COPY
# Reads saved json files

#import json so you dont have to run above cell if no new files needs saving
import json
import numpy as np
import os

#make all jsonfiles into keys


#create empty list
l = []
mock_well = []
OTP_well  = []
KIF11_well =[]
LIPA_well = []
PLK1_well = []
mock2_well= []
untr_well = []
label_X = []
all_plate_well = []

#loop through all keys (jsonfiles), load, add key to the empty list l
for k in Barcode_A:
    f4 = barcode_plate.loc[barcode_plate['Barcode']==k,'4fields'].values[0]
    #lb = k.split('-SA-')[1]+str((bool(f4))).replace('True','*').replace('False','')
    lb = k.split('-SA-')[1].split('-')[0][3:]
    label_X.append('')
    plate_data = np.zeros([16,24])
    pl   = barcode_plate.loc[barcode_plate['Barcode']==k, 'Plate'].values[0]
   
    if len(pl) != 0:
        with open('csv_files/A/'+pl.replace('MFGTMPcx7_','')+'_count.json','r') as f:
        
        ###Normalizing to avg of OTP wells
            json_str   = json.load(f)
            json_dict  = json.loads(json_str)
            plate_data = np.array(json_dict['data'])
             ##OTP


            ## This whole part is for adding 4 more field##
            P_ID  = pl
           
            ff =barcode_plate.loc[barcode_plate['Plate']==P_ID, '4fields'].values[0]
            fourfields = 'csv_files/A/4field/'+ff.replace('MFGTMPcx7_','')+'_count.json'
            if os.path.exists(fourfields):
                with open(fourfields,'r') as f:
                    second_json_4fields = json.load(f)
                    second_json_dict_4fields = json.loads(second_json_4fields)


                    second_plate_data = np.array(second_json_dict_4fields['data'])
                    plate_data = (plate_data+second_plate_data)*0.8


    OTP_per_plate =plate_data *np.array(OTP)
    otp_avg =  OTP_per_plate[np.nonzero(OTP_per_plate)].mean()
    plate_data = plate_data/otp_avg


    ##samples
    l.append(plate_data*np.array(samples))

    ##controls
    mock_well.append(plate_data*np.array(mock))
    OTP_well.append(plate_data*np.array(OTP))
    KIF11_well.append(plate_data*np.array(KIF11))
    LIPA_well.append(plate_data*np.array(LIPA))
    PLK1_well.append(plate_data*np.array(PLK1))
    mock2_well.append(plate_data*np.array(mock2))
    untr_well.append(plate_data*np.array(untr))

    all_plate_well.append(plate_data)
      
            

          
    

#label_X= plateNr
print(len(label_X))
       
#make list into array
array_all_plates = np.array(l)
mock_plates = np.array(mock_well)
OTP_plates = np.array(OTP_well)
KIF11_plates = np.array(KIF11_well)
LIPA_plates = np.array(LIPA_well)
PLK1_plates = np.array(PLK1_well)
untr_plates = np.array(untr_well)
mock2_plates = np.array(mock2_well)


all_plates = np.array(all_plate_well)


num_of_plates = len(Barcode_A)


all_for_plot_A = np.zeros([1,num_of_plates*384])
mock_for_plot_A = np.zeros([1,num_of_plates*384])
OTP_for_plot_A = np.zeros([1,num_of_plates*384])
KIF11_for_plot_A = np.zeros([1,num_of_plates*384])
LIPA_for_plot_A = np.zeros([1,num_of_plates*384])
PLK1_for_plot_A = np.zeros([1,num_of_plates*384])
untr_for_plot_A = np.zeros([1,num_of_plates*384])
mock2_for_plot_A = np.zeros([1,num_of_plates*384])

all_plate_for_plot_A = np.zeros([1,num_of_plates*384])

for plate_n in range(num_of_plates):
    wells = np.reshape(array_all_plates[plate_n,:,:],(1,384))
    all_for_plot_A[0,plate_n*384:(plate_n+1)*(384)]= wells

    mock_for_plot_A[0,plate_n*384:(plate_n+1)*(384)]= np.reshape(mock_plates[plate_n,:,:],(1,384))
    OTP_for_plot_A[0,plate_n*384:(plate_n+1)*(384)]= np.reshape(OTP_plates[plate_n,:,:],(1,384))
    KIF11_for_plot_A[0,plate_n*384:(plate_n+1)*(384)]= np.reshape(KIF11_plates[plate_n,:,:],(1,384))
    LIPA_for_plot_A[0,plate_n*384:(plate_n+1)*(384)]= np.reshape(LIPA_plates[plate_n,:,:],(1,384))
    PLK1_for_plot_A[0,plate_n*384:(plate_n+1)*(384)]= np.reshape(PLK1_plates[plate_n,:,:],(1,384))
    untr_for_plot_A[0,plate_n*384:(plate_n+1)*(384)]= np.reshape(untr_plates[plate_n,:,:],(1,384))
    mock2_for_plot_A[0,plate_n*384:(plate_n+1)*(384)]= np.reshape(mock2_plates[plate_n,:,:],(1,384))
    
    all_plate_for_plot_A[0,plate_n*384:(plate_n+1)*(384)]= np.reshape(all_plates[plate_n,:,:],(1,384)) 
    

#fig = plt.figure(figsize=(15,20))
#ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes

fig, ax1 = plt.subplots(1,1,figsize=(25,15))

print(len(all_for_plot_A[0]))

plt.scatter(np.arange(all_for_plot_A.shape[1]),all_for_plot_A[0],color='gray',marker='*')

plt.scatter(np.arange((OTP_for_plot_A.shape[1])),OTP_for_plot_A,color='blue')
plt.scatter(np.arange((mock_for_plot_A.shape[1])),mock_for_plot_A,color='lime')

plt.scatter(np.arange((PLK1_for_plot_A.shape[1])),PLK1_for_plot_A,color ='cyan')

plt.scatter(np.arange((LIPA_for_plot_A.shape[1])),LIPA_for_plot_A,color='orange')
plt.scatter(np.arange((KIF11_for_plot_A.shape[1])),KIF11_for_plot_A,color='m')

plt.scatter(np.arange((untr_for_plot_A.shape[1])),untr_for_plot_A,color='r')

plt.scatter(np.arange((mock2_for_plot_A.shape[1])),mock2_for_plot_A,color='y')


X      = np.arange(num_of_plates)*384

plt.vlines(x=X,ymin=0,ymax=40000,colors='purple', ls='--', lw=0.1)


ax1.set_xticks(X+192)

ax1.set_xticklabels(label_X, rotation=90)
plt.xlabel('Human genome')
plt.ylabel('Nuclei Count (Fold change)')
plt.rcParams.update({'font.size': 40})
plt.ylim([0.1,2])
plt.legend(['samples','OTP','mock','siPLK1','LIPA','KIF11','Untransfected','mock2'])
plt.title('Nuclei count normalized to average OTP')


fig, ax2 = plt.subplots(1,1,figsize=(10,5))
fig.set_size_inches((16,9))

sns.violinplot([OTP_for_plot_A[OTP_for_plot_A>0],mock_for_plot_A[mock_for_plot_A>0],
                mock2_for_plot_A[mock2_for_plot_A>0],untr_for_plot_A[untr_for_plot_A>0],LIPA_for_plot_A[LIPA_for_plot_A>0],
                PLK1_for_plot_A[PLK1_for_plot_A>0],
                KIF11_for_plot_A[KIF11_for_plot_A>0],all_for_plot_A[all_for_plot_A>0]],
                palette=['blue','lime','yellow','red','orange','cyan','magenta','gray'],width=1.7)
label_X = ['OTP','mock','mock2','Untransfected','LIPA','PLK1','KIF11','samples']
ax2.set_xticklabels(label_X, rotation=90)

plt.ylim([0.01,2])

plt.ylabel('Nuclei count (fold change)')
plt.title('Nuclei count normalized to average OTP')

np.save('A_normalized_count.npy',all_plate_for_plot_A,allow_pickle=True)
```
